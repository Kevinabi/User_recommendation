```python
ratings_with_name = ratings.merge(Books,on='ISBN')
ratings_with_name
```

```python
num_ratings = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>250
x
```

```python
freq_readers = x[x].index
freq_readers
```

```python
filter_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(freq_readers)]
filter_ratings
```

```python
y = filter_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
famous_books
```

```python
#final_ratings

final_ratings = filter_ratings[filter_ratings['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
#recomendation system
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_sc = cosine_similarity(pt)
```

```python
index = np.where(pt.index=='2nd Chance')[0][0]
index
pt.head(60)
```

```python
def recommendation_system(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_itm = sorted(list(enumerate(similarity_sc[index])),key= lambda x : x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_itm:
        item = []
        temp_df = Books[Books['Book-Title']==pt.index[i[0]]]
        item.extend( list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        
        data.append(item)
    print(index)    
    return data
```

```python
recommendation_system('1984')
```
